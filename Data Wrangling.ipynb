### **Descripción general**

```python
#Librerías

import pandas as pd
import numpy as np
import pingouin as pg
import matplotlib.pyplot as plt 
from scipy.stats import norm 
from scipy.stats import shapiro
from matplotlib.pyplot import FormatStrFormatter
```

```python
df = pd.read_csv('Practicas_encuesta.csv', encoding='utf-8')
```

```python
selected_df=df[["Matrícula","Semestre", "Terminal", "P1", "P2", "P3", "P4", "P5", "Total"]]
sorted_df=selected_df.sort_values(by="Semestre")
sorted_df.style.hide(axis="index")
```

```python
latex_code_df= sorted_df.to_latex()
print(latex_code_df)
```

```python
additional_metrics = pd.DataFrame({
    'P1': ['0 - 20', df['P1'].mean(), df['P1'].std(), df['P1'].skew(), df['P1'].kurt(), shapiro(df['P1'])[1], 'Sí' if shapiro(df['P1'])[1]>0.05 else 'No'],
    'P2': ['0 - 20', df['P2'].mean(),  df['P2'].std(), df['P2'].skew(), df['P2'].kurt(),shapiro(df['P2'])[1], 'Sí' if shapiro(df['P2'])[1]>0.05 else 'No'],
    'P3': ['0 - 20', df['P3'].mean(),  df['P3'].std(), df['P3'].skew(), df['P3'].kurt(),shapiro(df['P3'])[1], 'Sí' if shapiro(df['P3'])[1]>0.05 else 'No'],
    'P4': ['0 - 20', df['P4'].mean(),  df['P4'].std(), df['P4'].skew(), df['P4'].kurt(),shapiro(df['P4'])[1], 'Sí' if shapiro(df['P4'])[1]>0.05 else 'No'],
    'P5': ['0 - 20', df['P5'].mean(),  df['P5'].std(), df['P5'].skew(), df['P5'].kurt(),shapiro(df['P5'])[1], 'Sí' if shapiro(df['P5'])[1]>0.05 else 'No'],
    'Total': ['0 - 100', df['Total'].mean(),   df['Total'].std(), df['Total'].skew(), df['Total'].kurt(), shapiro(df['Total'])[1], 'Sí' if shapiro(df['Total'])[1]>0.05 else 'No'],
}, index=['Rango', 'Media', 'D. Estándar', 'Simetría', 'Curtosis', 'Test Shapiro', 'Dist. Normal'])
additional_metrics

```

```python
latex_code_ad= additional_metrics.to_latex()
print(latex_code_ad)
```

```python
grades_df = df[['P1','P2','P3','P4','P5','Total']]
grades_df.boxplot(showfliers=True, medianprops=dict(color="red"), showmeans=True, meanline=True, meanprops=dict(color='green'))
plt.ylim(-10, 110) 
plt.grid(False)
plt.title("Desempeño por pregunta")
plt.ylabel("Calificación")
plt.xlabel("Preguntas")
plt.show()
```

```python
# Histograma
plt.hist(df[["Total"]], bins=25, density=True, alpha=0.6, color='blue')
#plt.yticks(ticks=np.arange(0,0.12,0.02),labels=np.linspace(0,0.4,6))


# Acoplar a Distribución Normal
mu, std = norm.fit(df[["Total"]])
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)

# Gráficación
plt.plot(x, p, 'k', linewidth=2)
plt.title("Distribución de calificaciones")
plt.xlabel("Calificación")  
plt.ylabel("Densidad")
plt.show()
```
